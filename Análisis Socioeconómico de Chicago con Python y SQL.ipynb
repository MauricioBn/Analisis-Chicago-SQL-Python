```python
import sqlite3
conn = sqlite3.connect('socioeconomic.db')
cursor_obj = conn.cursor()
```

```python
#Crear tabla
table = """CREATE TABLE IF NOT EXISTS socioeconomic(CA INTEGER, community_area_name VARCHAR(50), percent_of_housing_crowded REAL, percent_households_below_poverty REAL, percent_aged_16_unemployed REAL, percent_aged_25_without_high_school_diploma REAL, percent_aged_under_18_or_over_64 REAL, per_capita_income_ REAL, hardship_index INTEGER);"""
cursor_obj.execute(table)
```
